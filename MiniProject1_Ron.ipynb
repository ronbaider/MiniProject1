```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
filename = 'laptop-price - dataset.csv'
# Using encoding='latin-1' to handle potential special characters often found in this dataset
df = pd.read_csv(filename, encoding='latin-1')

# Task 1: Plot the price of all laptops (Distribution)
plt.figure(figsize=(10, 6))
sns.histplot(df['Price (Euro)'], kde=True, bins=30)
plt.title('Distribution of Laptop Prices (Euro)')
plt.xlabel('Price (Euro)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.5)
plt.show()

# Task 2: Average price per company
avg_price_company = df.groupby('Company')['Price (Euro)'].mean().sort_values(ascending=False)

# Identify the company with the highest average
most_expensive_company = avg_price_company.idxmax()
highest_avg_price = avg_price_company.max()

print(f"Company with the highest average price: {most_expensive_company} ({highest_avg_price:.2f} Euro)")
print("\nAverage price per company:")
print(avg_price_company)

# Optional: Visualization for Average Price per Company
plt.figure(figsize=(14, 7))
sns.barplot(x=avg_price_company.index, y=avg_price_company.values, palette='viridis')
plt.title('Average Laptop Price by Company')
plt.xlabel('Company')
plt.ylabel('Average Price (Euro)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
